```python
plt.figure(figsize=(15,6))
sns.countplot(x='MMI', data = df, palette='hls')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(df['MMI'].value_counts(), labels=df['MMI'].value_counts().index, autopct="%1.1f%%" ,textprops= {'color':'black','weight':'bold','family':'serif'})
hfont={'fontname':'serif','weight':'bold'}
plt.title('MMI',size=20, **hfont)
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.histplot(df['Magnitude'],kde=True, palette='hls')
plt.xticks(rotation=90)
plt.show()
```
